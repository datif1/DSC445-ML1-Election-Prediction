### (1) Imports

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", None)
```

### (2) Load processed dataset

```python
df = pd.read_csv(
    "../data/processed/merged_county_dataset.csv",
    dtype={"county_fips": str},
    low_memory=False
)

df["county_fips"] = df["county_fips"].astype(str).str.zfill(5)

print(df.shape)
display(df.head())
```

### (3) Basic dataset overview

```python
print("Dataset Shape:")
print(df.shape)

print("\nColumn Count:")
print(len(df.columns))

print("\nData Types:")
display(df.dtypes.head(20))
```

### (4) Missing values analysis

```python
missing_values = (
    df.isnull()
    .sum()
    .sort_values(ascending=False)
)

missing_values = missing_values[missing_values > 0]

print("Columns with missing values:")
display(missing_values)
```

### (5) Basic statistics

```python
display(df.describe())
```

### (6) Check classification target balance

```python
print(df["party_winner"].value_counts())

sns.countplot(x="party_winner", data=df)

plt.title("Democrat vs Republican County Winners")
plt.xlabel("Party Winner (1 = Democrat, 0 = Republican)")
plt.ylabel("Number of Counties")

plt.show()
```

### (7) Analyze Democratic vote share

```python
plt.figure(figsize=(10, 5))

sns.histplot(df["dem_vote_share"], bins=30, kde=True)

plt.title("Distribution of Democratic Vote Share")
plt.xlabel("Democratic Vote Share")
plt.ylabel("Count")

plt.show()
```

### (8) Population distribution

```python
plt.figure(figsize=(10, 5))

sns.histplot(df["B01001_001E"], bins=50)

plt.title("County Population Distribution")
plt.xlabel("Population")
plt.ylabel("Count")

plt.show()
```

### (9) Median income distribution

```python
plt.figure(figsize=(10, 5))

sns.histplot(df["B19013_001E"], bins=40, kde=True)

plt.title("Median Household Income Distribution")
plt.xlabel("Median Household Income")
plt.ylabel("Count")

plt.show()
```

### (10) Democrat vs Republican income comparison

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    x="party_winner",
    y="B19013_001E",
    data=df
)

plt.title("Median Income by County Winner")
plt.xlabel("Party Winner (1 = Democrat, 0 = Republican)")
plt.ylabel("Median Household Income")

plt.show()
```

### (11) Poverty comparison by party

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    x="party_winner",
    y="B17001_001E",
    data=df
)

plt.title("Poverty-Related Population by County Winner")
plt.xlabel("Party Winner")
plt.ylabel("Population Count")

plt.show()
```

### (12) Correlation with vote share

```python
numeric_df = df.select_dtypes(include=np.number)

correlations = (
    numeric_df.corr()["dem_vote_share"]
    .sort_values(ascending=False)
)

display(correlations.head(15))
display(correlations.tail(15))
```

### (13) Correlation heatmap (top features)

```python
top_features = correlations.abs().sort_values(ascending=False).head(10).index

corr_matrix = df[top_features].corr()

plt.figure(figsize=(10, 8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Top Feature Correlation Heatmap")

plt.show()
```

### (14) Check duplicate counties

```python
print(
    "Duplicate county_fips:",
    df["county_fips"].duplicated().sum()
)
```

### (15) Final EDA summary checks

```python
print("Total counties:", len(df))

print("\nDemocrat-winning counties:")
print((df["party_winner"] == 1).sum())

print("\nRepublican-winning counties:")
print((df["party_winner"] == 0).sum())

print("\nAverage Democratic vote share:")
print(df["dem_vote_share"].mean())
```

# =======================================================
# Enhanced Dataset EDA (Feature Engineering Phase)
# =======================================================

### (16) Create a working copy

```python
eda_df = df.copy()
```

### (17) Create readable aliases

```python
eda_df["total_population"] = eda_df["B01001_001E"]

eda_df["median_income"] = eda_df["B19013_001E"]

eda_df["total_housing_units"] = eda_df["B25001_001E"]

eda_df["labor_force"] = eda_df["B23025_002E"]

eda_df["unemployed"] = eda_df["B23025_005E"]

eda_df["below_poverty"] = eda_df["B17001_002E"]

eda_df["bachelors_degree"] = eda_df["B15003_022E"]

eda_df["masters_degree"] = eda_df["B15003_023E"]

eda_df["professional_degree"] = eda_df["B15003_024E"]

eda_df["doctorate_degree"] = eda_df["B15003_025E"]

eda_df["white_population"] = eda_df["B02001_002E"]

eda_df["black_population"] = eda_df["B02001_003E"]

eda_df["asian_population"] = eda_df["B02001_005E"]
```

### (18) Create rate/percentage features

```python
# education rate
eda_df["higher_education_rate"] = (
    eda_df["bachelors_degree"] +
    eda_df["masters_degree"] +
    eda_df["professional_degree"] +
    eda_df["doctorate_degree"]
) / eda_df["total_population"]

# poverty rate
eda_df["poverty_rate"] = (
    eda_df["below_poverty"] /
    eda_df["total_population"]
)

# unemployment rate
eda_df["unemployment_rate"] = (
    eda_df["unemployed"] /
    eda_df["labor_force"]
)

# racial composition
eda_df["white_pct"] = (
    eda_df["white_population"] /
    eda_df["total_population"]
)

eda_df["black_pct"] = (
    eda_df["black_population"] /
    eda_df["total_population"]
)

eda_df["asian_pct"] = (
    eda_df["asian_population"] /
    eda_df["total_population"]
)
```

### (19) Check engineered features

```python
engineered_features = [
    "higher_education_rate",
    "poverty_rate",
    "unemployment_rate",
    "white_pct",
    "black_pct",
    "asian_pct"
]

display(
    eda_df[engineered_features].describe()
)
```

### (20) Compare education rate by party

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    x="party_winner",
    y="higher_education_rate",
    data=eda_df
)

plt.title("Higher Education Rate by Party Winner")
plt.xlabel("Party Winner (1 = Democrat, 0 = Republican)")
plt.ylabel("Higher Education Rate")

plt.show()
```

### (21) Compare poverty rate by party

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    x="party_winner",
    y="poverty_rate",
    data=eda_df
)

plt.title("Poverty Rate by Party Winner")
plt.xlabel("Party Winner")
plt.ylabel("Poverty Rate")

plt.show()
```

### (22) Compare unemployment rate by party

```python
plt.figure(figsize=(10, 6))

sns.boxplot(
    x="party_winner",
    y="unemployment_rate",
    data=eda_df
)

plt.title("Unemployment Rate by Party Winner")
plt.xlabel("Party Winner")
plt.ylabel("Unemployment Rate")

plt.show()
```

### (23) Compare race composition by party

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

sns.boxplot(
    x="party_winner",
    y="white_pct",
    data=eda_df,
    ax=axes[0]
)

axes[0].set_title("White Population %")

sns.boxplot(
    x="party_winner",
    y="black_pct",
    data=eda_df,
    ax=axes[1]
)

axes[1].set_title("Black Population %")

sns.boxplot(
    x="party_winner",
    y="asian_pct",
    data=eda_df,
    ax=axes[2]
)

axes[2].set_title("Asian Population %")

plt.tight_layout()
plt.show()
```

### (24) Correlation of engineered features

```python
selected_features = [
    "higher_education_rate",
    "poverty_rate",
    "unemployment_rate",
    "white_pct",
    "black_pct",
    "asian_pct",
    "median_income",
    "dem_vote_share"
]

corr_matrix = eda_df[selected_features].corr()

plt.figure(figsize=(10, 8))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap of Engineered Features")

plt.show()
```

### (25) Scatterplot: education vs Democratic vote share

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    x="higher_education_rate",
    y="dem_vote_share",
    data=eda_df
)

plt.title("Education Rate vs Democratic Vote Share")
plt.xlabel("Higher Education Rate")
plt.ylabel("Democratic Vote Share")

plt.show()
```

### (26) Scatterplot: income vs Democratic vote share

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    x="median_income",
    y="dem_vote_share",
    data=eda_df
)

plt.title("Median Income vs Democratic Vote Share")
plt.xlabel("Median Household Income")
plt.ylabel("Democratic Vote Share")

plt.show()
```

```python
eda_df["log_population"] = np.log1p(
    eda_df["total_population"]
)

plt.figure(figsize=(10, 5))

sns.histplot(
    eda_df["log_population"],
    bins=40
)

plt.title("Log-Transformed Population Distribution")

plt.show()
```
